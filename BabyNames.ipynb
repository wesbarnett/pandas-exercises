```python
def get_max_record(group):
    return group.loc[group.prop.idxmax()]
```

```python
# Most popular boys name in the record
get_max_record(boys)
```

```python
# Gets the most popular boys name each year
result = boys.groupby('year').apply(get_max_record)
```

```python
# proportion of most popular boy name over time
result.prop.plot()
```

```python
def get_quantile_count(group, quantile=0.5):
    group = group.groupby('soundex').sum()
    df = group.sort_values(by='prop', ascending=False)
    return df.prop.cumsum().searchsorted(quantile)[0]
```

```python
boy_ct = boys.groupby('year').apply(get_quantile_count)
```

```python
girl_ct = girls.groupby('year').apply(get_quantile_count,)
girl_ct.plot(label="girl")
boy_ct.plot(label="boy")
plt.legend()
```

```python
boy_ct = boys.groupby('year').apply(get_quantile_count, quantile=0.25)
girl_ct = girls.groupby('year').apply(get_quantile_count, quantile=0.25)
girl_ct.plot(label="girl")
boy_ct.plot(label="boy")
plt.legend()
```
